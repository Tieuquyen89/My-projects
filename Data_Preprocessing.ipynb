## Data pre-processing
In this step, we have to do data pre-processing to clean up data, remove duplicated, null values, and use regular expression to make it tidy & logic 

```python
#Import library to combine/merge csv files
import os, glob
import pandas as pd
import numpy as np
```

```python
path = "C:/Users/ADMIN/Desktop/Data Science/My projects/Job EDA"
all_files = glob.glob(os.path.join(path, "Career_Builder_*.csv"))
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
data_merged  = pd.concat(df_from_each_file, ignore_index=True)
data_merged.to_csv( "Career_Builder.csv")
```

```python
data=pd.read_csv('Career_Builder.csv')
data.info()
```

```python
# Handling Industry data by regular expression, create column of 'Industry_re'
import re
d=[]
for i in range(len(data['Industry'])):
    if (data.loc[i,'Industry']!='Not found'):
        chuoi = data.loc[i,'Industry']
        job_industry=re.search('\n(.*)\r',chuoi).group(1).replace(" ", "")
    else:
        job_industry='None'
    d.append({'Industry_re': job_industry})
    
industry=pd.DataFrame(d)
```

```python
industry.head()
```

```python
data = pd.concat([data, industry], axis=1, join='inner')
data.columns
```

```python
# Handling Experience data by regular expression, create column of 'Experience_re'
import re
d=[]
for i in range(len(data['Experience'])):
    chuoi = data.loc[i,'Experience']
    if (len(chuoi)>100):
        job_experience=re.search('\n(.*)\r',chuoi).group(1).replace(" ", "") 
    elif (len(chuoi))<50 and (len(chuoi))>10:
        job_experience=re.search('>(.*)<',chuoi).group(1)
    else:
        job_experience= 'fresher'
    d.append({'Experience_re': job_experience})
    
experience=pd.DataFrame(d)
```

```python
experience.head()
```

```python
data = pd.concat([data, experience], axis=1, join='inner')
data.columns
```

```python
# Handling Benefit data by regular expression, create column of 'Benefit_re'
import re
d=[]

for i in range(len(data['Benefit'])):
    chuoi = data.loc[i,'Benefit']
    if chuoi=='Not found':
        new_strings='None'       
    else:
        job_benefit=re.findall('</span>[\w ]{0,}</li>',str(chuoi))
        new_strings = []
        for string in job_benefit:
            new_string = string.replace("</span>", "")
            new_string = new_string.replace("</li>","").replace("['","")
            new_strings.append(new_string)
            
    d.append({'Benefit_re': new_strings})
benefit=pd.DataFrame(d)
```

```python
benefit.head()
```

```python
# Handling Description data by regular expression, create column of 'Description_re'
d=[]
for i in range(len(data['Description'])):
    chuoi = data.loc[i,'Description']
    if chuoi is None:
        new_strings='None'       
    else:
        job_description=re.findall('>(.*)>',chuoi)
        new_strings = []
        for string in job_description:
            new_string = (string.replace('<p>', '').replace('</p>','').replace('<strong>','').replace('</p','').replace('[]','').
            replace('</strong>','')).replace('[','').replace('<em>','').replace('</em>','').replace('<p','').replace('<br>','')
            new_strings.append(new_string)
    d.append({'Description_re': new_strings})   
description=pd.DataFrame(d)
```

```python
description.tail()
```

```python
# Handling Requirement data by regular expression, create column of 'Requirement_re'
d=[]
for i in range(len(data['Requirement'])):
    chuoi = data.loc[i,'Requirement']
    if chuoi is None:
        new_strings='None'       
    else:
        job_description=re.findall('>(.*)>',chuoi)
        new_strings = []
        for string in job_description:
            new_string = string.replace('[','').replace('<p>', '').replace('</p>','').replace('<strong>','').replace('</strong>','').replace('\xa0','').replace(']',"").replace('<em>','').replace('</em>','').replace('<p','').replace('<br/>','')
            new_strings.append(new_string)
    d.append({'Requirement_re': new_strings})   
requirement=pd.DataFrame(d)
requirement.tail()
```

```python
data = pd.concat([data,description,requirement,benefit], axis=1, join='inner') #, benefit
data.head()
```

```python
#Checking Null data
data.isnull().sum()
```

```python
data=data.drop(data.columns[[0,1,4,8,10,11,12,13]], axis=1)
data.info()
```

```python
data.to_csv('Final_Data_Career.csv')
```
