# Load data 

```python
# Import library 
import glob
import pandas as pd
import requests
import json
```

```python
path = r'C:\Users\ADMIN\Desktop\Assignment\Station_data' # my computer path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

station = pd.concat(li, axis=0, ignore_index=True)
station.shape
```

```python
station.head()
```

```python
station=station.drop(station.columns[0],axis=1)
pd.set_option('display.max_rows', None)
station.head()
```

```python
station=station.drop_duplicates()
station.shape
```

```python
station=station.dropna(how='any')
station.shape
```

# API get the Long & Lat 

```python
for i, row in station.iterrows():
    apiAddress = str(station.at[i,'Address'])+','+str(station.at[i,'Post_code'])+','+str(station.at[i,'City'])
    
    parameters = {
        "key": "RbHtUg1m5l3tZRS9NG4h19pQAgr909YS",
        "location": apiAddress
    }

    response = requests.get("http://www.mapquestapi.com/geocoding/v1/address", params=parameters)
    #print(i," ",response)
    data = response.text
    dataJ = json.loads(data)['results']
    lat = (dataJ[0]['locations'][0]['latLng']['lat'])
    lng = (dataJ[0]['locations'][0]['latLng']['lng'])

  
    station.at[i,'lat'] = lat
    station.at[i,'lng'] = lng
    
station.to_csv('station_geo.csv')
```
